```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from nltk.tokenize import sent_tokenize

text = testd['text'][0]

per = 0.3

# Creating text parser using tokenization
parser = PlaintextParser.from_string(testd['text'][0],Tokenizer("english"))

p = np.ceil(len(sent_tokenize(text)) * per)
if(p<3 and len(sent_tokenize(text))>3):
        p = np.ceil(len(sent_tokenize(text)) * per * 2)#for small text

from sumy.summarizers.text_rank import TextRankSummarizer

# Summarize using sumy TextRank
summarizer = TextRankSummarizer()
summary =summarizer(parser.document,p)

text_summary=""
for sentence in summary:
    text_summary+=str(sentence)

print(text_summary)
```
