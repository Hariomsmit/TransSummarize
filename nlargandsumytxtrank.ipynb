```python
#test set extractive by sumy 
def sumy_sum(text,per = 0.2):
    from sumy.parsers.plaintext import PlaintextParser
    from sumy.nlp.tokenizers import Tokenizer
    from sumy.summarizers.text_rank import TextRankSummarizer
    from nltk.tokenize import sent_tokenize

    p = np.ceil(len(sent_tokenize(text)) * per)

    parser = PlaintextParser.from_string(text,Tokenizer("english"))

    summarizer = TextRankSummarizer()
    summary = summarizer(parser.document,p)

    text_summary = ""
    for sentence in summary:
        text_summary+=str(sentence)
    
    return text_summary
```

```python
doc = nlp(text)
doc
```

```python
tokens = [token.text for token in doc]
print(tokens)
```

```python
word_freq = {}
for word in doc:
    if word.text.lower() not in l_stopwrds:
        if word.text.lower() not in puncts:
            if word.text not in word_freq.keys():
                word_freq[word.text] = 1
            else:
                word_freq[word.text] += 1
print(word_freq)
```

```python
max_freq = max(word_freq.values())
max_freq
```

```python
for word in word_freq.keys():
    word_freq[word] = word_freq[word]/max_freq
    print(word_freq)
```

```python
sent_tokens = [sent for sent in doc.sents]
print(sent_tokens)
```

```python
sent_scores = {}
for sent in sent_tokens:
    for word in sent:
        if word.text.lower() in word_freq.keys():
            if sent not in sent_scores.keys():
                sent_scores[sent] = word_freq[word.text.lower()]
            else:
                sent_scores[sent] += word_freq[word.text.lower()]
sent_scores
```

```python
select_len = int(len(sent_tokens)*0.3)
select_len
```

```python
summary = nlargest(select_len, sent_scores, key = sent_scores.get)
summary
```

```python
f_summary = [word.text for word in summary]
summary = ' '.join(f_summary)
```
